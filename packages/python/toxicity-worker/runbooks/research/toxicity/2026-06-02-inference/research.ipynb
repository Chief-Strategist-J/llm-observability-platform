# Toxicity Worker Inference Research
This notebook demonstrates how to load the local `DetoxifyOnnxAdapter` and run inference on sample texts to verify toxicity scoring.

```python
import sys
import os
sys.path.append(os.path.abspath('../../../../src'))

from infra.adapters.detoxify_onnx_adapter import DetoxifyOnnxAdapter
from core.domain.service import score_toxicity
```

Initialize the adapter with the default unitary/toxic-bert model:

```python
adapter = DetoxifyOnnxAdapter()
print(f"Loaded model: {adapter.model_id}")
```

Tokenize and score a standard sample sentence (under 510 tokens):

```python
sample_text = "This is an awesome day and everything is working perfectly!"
token_ids = adapter.tokenize(sample_text)
print(f"Token IDs: {token_ids}")

result = score_toxicity(
    scorer=adapter,
    token_ids=token_ids,
    trace_id=None,
    span_id=None
)
print("\nSingle Pass Results:")
for field, val in result.scores.__dict__.items():
    print(f"  {field}: {val:.4f}")
```

Perform dual pass scoring for a long response (exceeding 510 tokens):

```python
long_text = "friendly greeting " * 550
long_token_ids = adapter.tokenize(long_text)
print(f"Long token length: {len(long_token_ids)}")

long_result = score_toxicity(
    scorer=adapter,
    token_ids=long_token_ids,
    trace_id=None,
    span_id=None
)
print("\nDual Pass Results:")
print(f"  Strategy used: {long_result.long_response_strategy}")
for field, val in long_result.scores.__dict__.items():
    print(f"  {field}: {val:.4f}")
```
